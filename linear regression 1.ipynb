```python
q1=14.114
q3=67.956
iqr=q3-q1
lf=q1-1.5*iqr
uf=q3+1.5*iqr
print("uf",uf)
```

here we r keeping the value as same the one which is greater than uf will be removed and rest fo the things will be the same


```python
df["Sales_in_thousands"]=np.where(df["Sales_in_thousands"]>uf,uf,df["Sales_in_thousands"])
```

```python
plt.boxplot(df['Sales_in_thousands'])
plt.show()
```

```python
for i in df.select_dtypes([float,int]).columns:
    plt.boxplot(df[i])
    plt.title(i)
    plt.show()
```

```python
for i in df.select_dtypes([float,int]).columns:
  q1=df[i].quantile(0.25)
  q3=df[i].quantile(0.75)
  iqr=q3-q1
  lf=q1-(1.5*iqr)
  uf=q3+(1.5*iqr)
  df[i]=np.where(df[i]>uf,uf,df[i])
  df[i]=np.where(df[i]<lf,lf,df[i])
```

```python
x=df.drop(columns=["Sales_in_thousands", "Vehicle_type", "Latest_Launch"])
y=df["Sales_in_thousands"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error,root_mean_squared_error
```

```python
print("MSE",mean_squared_error(y_test,lr.predict(x_test)))
print("RMSE",root_mean_squared_error(y_test,lr.predict(x_test)))
```

```python
actual=y_train
pred=lr.predict(x_train)
```

```python
error=actual - pred
sq_error=np.square(error)
MSE=np.mean(sq_error)
print(MSE)
```
